```python
class Student():
    def __init__(self,age):
        self.course=1
        self.age=age
    
    def next_course(self):
        print("Текущий курс", self.course)
        self.course +=1
        if self.course == 6:
            print("Обучение закончено")
                 
s1 = Student(18)
for i in range(5):
    s1.next_course()
```
