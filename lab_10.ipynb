Представити постачання 5 різних товарів на склади (створити 5 екземплярів класу) різного типу.
Застосувати метод зміни кількості товарів на складі
Роздрукувати залишки на складі
Провести інтерпретацію.

```python
class Sklad:
    def __init__(self, name="Product", type="Default",code="0", amount=10):
        self.name, self.type, self.code, self.amount=name, code, type, amount
        
    def setcode(self,code):
        if code==1:
            self.type="Склад продовольчих товаров"
        elif code ==2:
            self.type="Склад продуктов"
        elif code ==3:
            self.type="Склад електроники"
        elif code ==4:
            self.type="Скад буд.материалов"
        elif code ==5:
            self.type="Склад одежды"
            
    def setamount(self, set_amount):
        self.amount=set_amount
    
    def prod_reduce(self):
        self.amount-=1
        if self.amount<0:
            self.amount=0

t1=Sklad()
t1.setcode(1)

t2=Sklad()
t2.setcode(2)

t3=Sklad()
t3.setcode(3)

t4=Sklad()
t4.setcode(4)

t5=Sklad()
t5.setcode(5)

t1.setamount(56)
t2.setamount(34)
t3.setamount(432)
t4.setamount(234)
t5.setamount(23)

t2.prod_reduce()
t2.prod_reduce()
t2.prod_reduce()

print("1 - Склад продовольчих товаров:", t1.amount,"\n2 - Склад продуктов:", t2.amount, "\n3 - Склад електроники:", t3.amount, "\n4 - Скад буд.материалов:", t4.amount, "\n5 - Склад одежды:", t5.amount)
```
